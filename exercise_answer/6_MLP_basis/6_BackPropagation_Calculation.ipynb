## 연습문제 3

아래 그림은 sigmoid 표현한 graph 입니다. 


x = 3 일때 
$\frac{\partial L}{\partial x}$ 을 구해 answer3 에 넣어주세요.<br>


![Imgur](https://i.imgur.com/MdxPuau.png)

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer3`에 정답을 담아주세요

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# 정답을 입력해 주세요
answer3 = sigmoid(3)*(1-sigmoid(3))
```

```python
from numpy.testing import assert_almost_equal
assert 'answer3' in globals(), "answer3에 값을 담았는지 확인해주세요"

A = 0.045176659730912
assert_almost_equal(answer3, A, decimal=5,
                    err_msg="정답이 {} 인지 확인해 주세요".format(A))
print("정답입니다!!!!")
```

## 연습문제 4
아래 그림을 이용해서 x = 3 , y=1, Loss=2.6 일때 <br> $\frac{\partial loss}{\partial W^0_{00}}$  을 구해 answer4 에 넣어주세요 


![Imgur](https://i.imgur.com/KPr1p67.png)





#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer4`에 정답을 담아주세요

```python
x = 3
y = 1
loss = 2.6

w0 = 0.12
w1 = 0.07

# forward
z = x * w0
a = sigmoid(z)
logits = a * w1
pred = sigmoid(logits)

# backward
dL_dlogits = pred - y
dL_da = dL_dlogits * w1
da_dz = a*(1-a)
dz_dw = x
dL_dw0 = dL_da * da_dz * dz_dw

answer4 = dL_dw0
```

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

```python
from numpy.testing import assert_almost_equal
assert 'answer4' in globals(), "answer4에 값을 담았는지 확인해주세요"

A = -0.024893594805934935
assert_almost_equal(answer4, A, decimal=5,
                    err_msg="정답이 {} 인지 확인해 주세요".format(A))
print("정답입니다!!!!")
```

* 아래에 답을 서술하여 주세요
* `answer5`에 정답을 담아주세요

```python
# 합성함수 미분 : dL_dw0 = dL_dpred * dpred_dlogit * dlogit_da * da_dz * dz_dw0
x = -3
y = 1
loss = -2.6
w0 = 0.12
w1 = 0.07

# forward
z = x * w0
a = np.where(z > 0, z, 0)
logits = a * w1
pred = sigmoid(logits)

# backward
dL_dlogits = pred - y
dlogits_da = w1
da_dz = np.where(z > 0, 1, 0)
dz_dw0 = x
dL_dw0 = dL_dlogits * dlogits_da * da_dz * dz_dw0

answer5 = dL_dw0
```

![Imgur](https://i.imgur.com/JQTcwcP.png)

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer6`에 정답을 담아주세요

```python
x = np.array(3.,ndmin=2)
y = np.array(1.,ndmin=2)
loss = 2.6
w0 = np.array([[0.12, -0.23]])
w1 = np.array([[0.07], [0.16]])

# forward
z = x * w0
a = sigmoid(z)
logits = np.matmul(a, w1)
pred = sigmoid(logits)

# backward
dL_dlogits = pred - y
dlogits_da = w1
da_dz = a * (1-a)
dz_dX = w0
dL_da = np.matmul(dL_dlogits, w1.T)
dL_dz = dL_da * da_dz
dL_dw0 = np.matmul(dL_dz, w0.T)

answer6 = dL_dw0
```

## 연습문제 7

$\frac{\partial Loss}{\partial W^1} $ 을 구해서 answer7 에 넣어주세요 

![Imgur](https://i.imgur.com/5Ap1SD4.png)

```python
# 우선 실행해주세요
X = np.array([[9., 3.],
             [3., 7.],
             [8., 4.],
             [2., 3.]])

Y = np.array([[1.],
             [0.],
             [1.],
             [0.]])


W1 = np.array([[0.3, -0.7],
               [0.6, 0.4]])

B1 = np.array([[0., 0.1]])



W2 = np.array([[0.4] ,[0.6]])

B2 = np.array([[0.7]])
```

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer7`에 정답을 담아주세요

```python
# minimal answer

# forward
z = np.matmul(X, W1) + B1
a = np.where(z > 0, z, 0)
logits = np.matmul(a, W2) + B2
pred = sigmoid(logits)

# backward
delta = pred-Y # sigmoid with cross entropy
delta = np.matmul(delta, W2.T)
answer7 = np.matmul(X.T, np.where(delta > 0, 1, 0))
```

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

```python
from numpy.testing import assert_array_almost_equal
assert 'answer7' in globals(), "answer7 값을 담았는지 확인해주세요"

A = np.array([
    [5.,5.],
    [10.,10.]
])

assert_array_almost_equal(answer7, A, decimal=5,
                          err_msg="정답이 {} 인지 확인해 주세요".format(A))
print("정답입니다!!!!")
```
