## 문제 1. 공격횟수 `얍` 세기

* `while`과 `if`를 통해, 선수의 기합 중 공격 소리인 `얍`의 횟수를 세어주세요

`````python
sound = ['얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', 
 '앗', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍']
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과값은 `answer1`에 담아주세요

`````python
cnt = 0

idx = 0
while idx < len(sound):
    if sound[idx] == '얍':
        cnt += 1
        
    idx += 1
    
answer1 = cnt
`````

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

`````python
assert 'answer1' in globals(), "결과를 answer1 변수에 담아주세요!, answer1 값을 보고 평가합니다."

if (answer1 == 45):
    print("정답입니다!!!!")
else:
    raise ValueError("45가 나와야합니다. 다시 확인해주세요!")
`````

## 문제 2. 승점 세기

* 선수가 공격을 할 때, 승점이 1점 올라가고, 선수가 공격을 당할 때, 승점이 1점 내려간다고 할 때, 우리의 선수는 몇 점을 얻었는지 계산해주세요

`````python
sound = ['얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', 
 '앗', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍']
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과값은 `answer2`에 담아주세요

`````python
score = 0 

idx = 0
while idx < len(sound):
    if sound[idx] == '얍':
        score += 1
    else:
        score -= 1
        
    idx += 1

answer2 = score
`````

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

`````python
assert 'answer2' in globals(), "결과를 answer2 변수에 담아주세요!, answer2 값을 보고 평가합니다."

if (answer2 == 40):
    print("정답입니다!!!!")
else:
    raise ValueError("40이 나와야합니다. 다시 확인해주세요!")
    
`````

## 문제 3. 첫 공격 횟수 세기

* 선수가 첫 공격 당하기 전까지(==처음 앗 소리가 나오기 전까지) 몇 번이나 공격을 했는지 계산해주세요

`````python
sound = ['얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', 
 '앗', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍']
`````

* 아래에 답을 서술하여 주세요
* 결과값은 `answer3`에 담아주세요

`````python
cnt = 0

idx = 0
while idx < len(sound):
    if sound[idx] == '앗':
        break
    
    cnt += 1
    idx += 1
    
answer3 = cnt
`````

#### 정답 확인

`````python
assert 'answer3' in globals(), "결과를 answer3 변수에 담아주세요!, answer3 값을 보고 평가합니다."

if (answer3 == 15):
    print("정답입니다!!!!")
else:
    raise ValueError("15가 나와야합니다. 다시 확인해주세요!")
    
`````

## 문제 4. 최대 연속 공격 횟수 세기

* 선수가 가장 길게 연속으로 공격한 횟수(얍 전까지 최대 공격 횟수)가 얼마인지 계산해주세요

예시) 
````python
['얍','얍','앗','얍','얍','얍','얍','앗','얍','얍','얍']
````
위에서 연속 공격은 아래와 같다.
* 첫번째 연속 공격 : ['얍','얍']
* 두번째 연속 공격 : ['얍','얍','얍',얍']
* 세번째 연속 공격 : ['얍','얍','얍]

이 중 2번째 연속 공격의 숫자 4가 제일 크므로, 정답은 4이다.

`````python
sound = ['얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', 
 '앗', '얍', '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '앗', '얍', '얍', '얍', '얍', '얍', '얍', '얍', 
 '얍', '얍', '얍', '얍', '얍', '얍', '얍', '앗', '얍', '얍']
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과값은 `answer4`에 담아주세요

`````python
cnt = 0
local_cnt = 0

i = 0
while i < len(sound):
  
    if sound[i] == '앗':
        cnt = max(cnt, local_cnt)
        local_cnt = 0
    else:
        local_cnt = local_cnt + 1
    
    i = i + 1
    
answer4 = max(cnt, local_cnt)
`````

#### 정답 확인
* 정답을 입력한 후, 아래를 실행시키면 정답인지 확인합니다.

`````python
assert 'answer4' in globals(), "결과를 answer4 변수에 담아주세요!, answer4 값을 보고 평가합니다."

if (answer4 == 15):
    print("정답입니다!!!!")
else:
    raise ValueError("15가 나와야합니다. 다시 확인해주세요!")
    
`````
