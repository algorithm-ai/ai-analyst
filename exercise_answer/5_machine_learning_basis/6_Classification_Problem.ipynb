## 문제 4. Logistic gradient 계산하기 

텐서플로우를 이용해 -10부터 10까지  0.1 단위로 logitstic 
함수를 미분해 시각화 해보세요



`````python
# 우선 실행해 주세요.
ps = np.arange(-10, 10, 0.1)
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요


`````python
def sigmoid(p):
    return 1/(1+np.exp(-p))
`````

`````python
ys = list(map(lambda p : sigmoid(p)*(1-sigmoid(p)), ps))
`````

`````python
plt.plot(ps, ys)
`````

## 문제 5. 엔트로피 계산하기

오늘 일어날 사건은 A, B, C, D 4가지 입니다. 

그리고 각 일어날 확률은 아래와 같습니다. 


A 사건은 일어날 확률이 0.7 입니다. <br>
B 사건은 일어날 확률이 0.26 입니다. <br>
C 사건은 일어날 확률이 0.02 입니다. <br>
D 사건은 일어날 확률이 0.02 입니다. <br>

오늘 일어날 사건의 엔트로피를 텐서플로우를 이용해 구해주세요.

`````python
# 우선 실행해 주세요.
ps = [0.7, 0.26, 0.02, 0.02]
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요.
* `answer5`에 답을 입력해 주세요.


`````python
P = tf.constant(ps)
I = -tf.log(ps)

sess = tf.Session()
sess.run(tf.reduce_sum(I*P))
`````

#### 정답 확인 

정답 :  0.7563925


## 문제 6. Cross - Entropy 계산하기

아래 경우 Cross-entropy 을 구해주세요.

|정답  | 0   | 1   |
|-----|-----|-----|
|예측값| 0.3 | 0.7 |

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer6`에 답을 입력해 주세요.


`````python
def binary_cross_entropy(p, p_hat):
    
    return tf.cond(tf.cast(p, tf.bool), lambda: -tf.log(p_hat), lambda: -tf.log(1-p_hat))
`````

`````python
data1 = [1, 0.7]

op = binary_cross_entropy(*data1)
tf.Session().run(op)
`````

`````python
data2 = [0, 0.3]

op = binary_cross_entropy(*data2)
tf.Session().run(op)
`````

#### 정답 확인 

정답 :  0.35667497
