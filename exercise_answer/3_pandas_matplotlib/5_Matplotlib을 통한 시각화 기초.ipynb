```python
# 우선 실행해 주세요
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import matplotlib.font_manager as fm

# 2. 나눔 폰트의 위치 가져오기 
system_font = fm.findSystemFonts() # 현재 시스템에 설치된 폰트
print(system_font)
nanum_fonts = [font for font in system_font if "NanumBarunGothic.ttf" in font]
font_path = nanum_fonts[0] # 설정할 폰트의 경로

# 3. 나눔 폰트로 설정하기
font_name = fm.FontProperties(fname=font_path, size=10).get_name()
plt.rc("font",family=font_name)

# 4. 폰트 재설정하기
fm._rebuild()

# 5. (optional) minus 기호 깨짐 방지
mpl.rcParams['axes.unicode_minus'] = False
```

## 문제 1. `sin` Plot 그리기
* sin_x의 값을 아래와 같은 형태로 그려주세요

![Imgur](https://i.imgur.com/zJDKEgS.png)
**조건**

1. 제목은 `사인 함수`로 해주세요.
2. x축의 범위는 `0~6*np.pi`로 해주세요.
3. y축의 범위는 `-2~2`로 해주세요.
4. x축의 이름은 `xs`로 해주세요.
5. y축의 이름은 `sin_x`로 해주세요.

```python
# 우선 실행해주세요
xs = np.arange(0,6*np.pi,np.pi/18)
sin_x = np.array([np.sin(x) for x in xs])
```

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과 그림이 위와 동일하도록 만들어 주세요

```python
# 답안지를 작성해 주세요.
plt.plot(xs, sin_x)
plt.xlim((0, 6*np.pi))
plt.ylim((-2,2))
plt.xlabel('xs')
plt.ylabel('sin_x')

plt.show()
```

## 문제 2. `sin&cos` Plot 그리기
* 두 주기 함수의 값을 아래와 같은 형태로 그려주세요
![Imgur](https://i.imgur.com/h45qS3l.png)
**조건**

1. 제목은 `주기 함수`로 해주세요.
2. x축의 범위는 `0~6*np.pi`로 해주세요.
3. y축의 범위는 `-2~2`로 해주세요.
4. x축의 이름은 `xs`로 해주세요.
5. y축의 이름은 `ys`로 해주세요.
6. 범례를 `sin`,`cos`으로 추가해주세요.
7. 지침선을 `[0,pi,2*pi,3*pi,4*pi,5*pi,6*pi]`로 표시해주세요.

```python
# 우선 실행해주세요
xs = np.arange(0,6*np.pi,np.pi/18)
sin_x = np.array([np.sin(x) for x in xs])
cos_x = np.array([np.cos(x) for x in xs])
```

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 결과 그림이 위와 동일하도록 만들어 주세요

```python
# 답안지를 작성해 주세요.
plt.plot(xs, sin_x)
plt.plot(xs, cos_x)

plt.xlim((0, 6*np.pi + 0.001))
plt.ylim((-2,2))

l = ['0','pi','2*pi','3*pi','4*pi','5*pi','6*pi']
ticks = (*plt.xlim(), np.pi)
plt.xticks(np.arange(*ticks), l)
plt.yticks([-1, 0, 1])

plt.xlabel('xs')
plt.ylabel('ys')
plt.legend(['sin', 'cos'])

plt.show()
```
