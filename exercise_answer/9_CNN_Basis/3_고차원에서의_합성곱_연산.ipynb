### (1) 3차원 데이터 구성하기

아래와 같이 3차원으로 구성된 입력 행렬을 구성하였습니다.<br>

```python
in0 = np.array([
    [1,4,2,0],
    [2,3,1,0],
    [3,1,2,3],
    [4,3,2,1]
])
in1 = np.array([
    [1,7,2,1],
    [3,2,1,9],
    [0,1,5,1],
    [4,6,2,4]
])
in2 = np.array([
    [1,5,2,1],
    [3,2,1,9],
    [0,1,3,1],
    [4,6,2,7]
])
inputs = np.stack([in0, in1, in2], axis=-1)
print("입력값의 형태 (H, W, C) :({},{},{})".format(*inputs.shape))
```

### (2) Filter 구성하기

여기서 중요한 것은 입력값의 채널 수(C)와 필터의 채널 수(C)가 동일해야 한다는 것입니다.<br>
입력값의 채널에 따라 채널 별 필터가 존재하고, 각 채널 별로 합성곱 연산을 진행한 후, 하나의 값으로 출력됩니다.<br>

```python
f0 = np.array([
    [2,0,1],
    [0,1,2],
    [1,0,2]
])
f1 = np.array([
    [0,1,3],
    [2,1,3],
    [4,1,2]
])
f2 = np.array([
    [3,2,1],
    [2,2,3],
    [0,0,1]
])
filter = np.stack([f0, f1, f2], axis=-1)
print("Filter의 형태 (H, W, C) : ({},{},{})".format(*filter.shape))
```

### (3) 합성곱 연산 진행하기

padding은 valid, stride은 1로 두고 연산하면 아래와 같습니다.

```python
outputs = np.zeros([2,2])# fix me # 
for i in range(2):
    for j in range(2):
        result = np.sum(filter * inputs[i:i+3, j:j+3])#fix me# 
        outputs[i,j] = result
        print("({},{}) 번째 값 : {}".format(i,j,result))
outputs
```

## 1. 2차원 데이터로 풀링 연산 수행하기
---

* 아래의 2차원 데이터를 통해, 풀링 연산이 어떤 식으로 진행되는지 살펴보도록 하겠습니다.

### (1) 2차원 데이터 구성하기

```python
values = np.array([
    [1,3,2,1],
    [2,9,1,1],
    [1,3,2,3],
    [2,6,1,2]
])
values
```

### (2) 풀링 연산 수행하기

아래는 2x2 Max Pooling을 stride 2로 처리하는 결과입니다.<br> 
최대 풀링은 합성곱 연산와 같이, 특정 패치 단위로 입력값을 스캔하면서<br>
처리하게 됩니다.

```python
filters = 2
strides = 2

n = int((4+2*1-2)/2)
result = np.zeros([n,n]) # pooling 후 결과물의 shape 을 적어주세요 
for i in range(n):
    for j in range(n):
        patch = values[i*strides:i*strides+n, j*strides:j*strides+n]
        result[i,j] = np.max(patch)
result
```

Pooling Layer는 이미지에서 Sampling하는 것과 효과와 동일합니다.<br>
이렇게 될 경우, 이미지가 약간 이동하더라도 신경망이 영향을 덜 받게 됩니다. <br>

이는 위치 불변성(Location Invariance)이라는 이슈로,<br> 
모델이 견고하게 되는데에 중요한 특성 중 하나입니다.

## 4. tf.nn.pooling 을 사용해 3차원 영상 Pooling  수행하기 
---

tf.nn.conv2d 을 이용해 어떻게 convoluiton 연산을 수행하는지 알아봅니다. 

tensorflow 에서 제공하는 pooling 은 2가지 종류가 있습니다. 

- max pooling : [tf.nn.max_pool](https://www.tensorflow.org/api_docs/python/tf/nn/max_pool)
- average pooling : [tf.nn.avg_pool](https://www.tensorflow.org/api_docs/python/tf/nn/avg_pool)



```python
from sklearn.datasets import load_sample_images
dataset = load_sample_images()    
first_img_data = dataset.images[1] 

plt.imshow(first_img_data)
plt.title('background')
plt.show()
```

```python
# normalize
first_img_data = first_img_data/255.#fix me #
# first_img_data to tensor
tf_dataset = tf.constant(first_img_data)

# pooling kernel shape 10, 10
# pooling stride shape 2, 2

tf_dataset = tf.reshape(tf_dataset, [1, *tf_dataset.shape])
maxpool_output = tf.nn.max_pool(tf_dataset, [1,10,10,1], [1,2,2,1], 'SAME')#fix me#
avgpool_output = tf.nn.avg_pool(tf_dataset, [1,10,10,1], [1,2,2,1], 'SAME')#fix me#

sess = tf.Session()
# sess.run(tf.global_variables_initializer())
maxpool_output_, avgpool_output_ = sess.run([maxpool_output, avgpool_output])
maxpool_output_ = np.squeeze(maxpool_output_)
avgpool_output_ = np.squeeze(avgpool_output_)

fig, axes = plt.subplots(1,2)

# axes0 : max pooling output , axes1 : max pooling output
axes[0].imshow(maxpool_output_)
axes[0].set_title('max')
axes[1].imshow(avgpool_output_)
axes[1].set_title('avg')
plt.show()
```
