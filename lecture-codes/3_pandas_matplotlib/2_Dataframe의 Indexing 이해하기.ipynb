## 1. 열 방향으로 조작하기

---

+ 우리는 열의 이름을 통해 열을 가져오고, 변경하고, 제거할 수 있습니다.<br>



### (1) 특정 열 가져오기

`Dataframe`에서 하나의 열을 선택하면, `Series`로 반환하게 됩니다.<br>
`dict`처럼 대괄호([])을 이용하거나<br>
`class`의 attribute처럼 `.<열이름>`을 이용하면 됩니다.

```python
# 국어 성적 가져오기
df['국어'] 
```

```python
# 학생 이름 가져오기
df.이름 
```

```python
# 복수개의 열 가져오기
df[["국어","영어","수학"]] 
```

### (2) 새로운 열 추가하기

* `Dataframe` 내 없는 열의 이름에 값을 넣으면, 새로운 열이 추가됩니다.

```python
# 학생의 시험 통과 유무에 대한 열을 추가
df["pass"] = True
df
```

* 값이 1개이면 일괄적으로 들어갑니다.Numpy의 Broadcasting과 유사하게 동작합니다.

```python
# pass에 대한 열을 각각 추가
df["pass"] = [True,True,False,False,
              True,True,True]
df
```

* Python List나 np.Array로 들어갈 수 있습니다.

### (3) 열 별로 계산하기

* Numpy.array의 계산과 마찬가지로, 열 별로 더하고 곱하는 등의 연산을 할 수 있습니다.

```python
# 국영수 평균 성적
df["국영수"] = (df.국어 + df.영어 + df.수학)/3 
df
```

* `numpy.array`와 마찬가지로, 원소 간에 매칭되어 연산이 동작합니다.

```python
# 국영수 기준 80점이 넘는 경우에만 시험에서 통과
df["통과"] = df["국영수"] > 80
df
```

### (4) 열 삭제하기

열을 삭제하기 위해서는 `del`키워드 혹은 `drop`메소드를 이용할 수 있습니다.

```python
del df["국영수"]
df
```

```python
df = df.drop(columns=["pass","통과"])
df
```

## 2. 행 방향으로 조작하기

---

+ 우리는 행의 인덱싱을 통해 행을 가져오고, 변경하고, 제거할 수 있습니다.<br>



### (1) 특정 행 가져오기

`Dataframe`에서 하나의 행을 선택하면, `Series`로 반환하게 됩니다.<br>
행을 가져오는 방법으로는 인덱스 이름으로 가져오는 `.loc()`과,
순서로 가져오는 `.iloc()`이 있습니다.<br>


```python
# 행의 index 변경하기
df.index = ["a","b","c","d","e","f","g"]
df
```

`loc`는 행의 이름을 통해 가져오는 방식입니다.

```python
# a행 가져오기
df.loc["a"]
```

`iloc`은 행의 **순서**를 통해 가져오는 방식입니다.<br>
numpy array 인덱싱과 유사하게, **위치**를 기반으로 가져옵니다.

```python
# 0번째 행 가져오기
df.iloc[0]
```

```python
# 복수개의 행 가져오기
df.loc["a":"c"]
```

slicing의 경우 마지막 원소도 포함한다는 점에서 <br>
numpy array의 indexing과 미묘하게 다릅니다.<br>
이는 이후 실수를 유발할 수 있는 부분이 됩니다.

```python
# 행과 열의 조합으로 가져오기
df.loc[["a","b"],
       ["이름","국어","영어","수학"]]
```

```python
df.loc[:,"국어"] # == df.국어
```

### (2) 새로운 행 추가하기

* `Dataframe` 내 없는 행의 이름에 값을 넣으면, 새로운 행의 값이 추가됩니다.

```python
# h 행에 추가
df.loc["h"] = ["4반","은진",95,83,85,87,80]
df
```

```python
# 마지막 행에 붙이고 싶다면, 아래와 작성하면 됩니다.
df.append({
    "반" : "4반",
    "이름" : "은효",
    "국어" : 95,
    "영어" : 70,
    "수학" : 82,
    "사회" : 91,
    "과학" : 75
},ignore_index=True)
```

### (3) 행 별로 계산하기 

* 열 방향과 유사한 방식으로 계산할 수 있습니다.

```python
# 승기와 호동의 성적의 차이
df.loc["a","국어":"과학"] - df.loc["b","국어":"과학"]
```

### (4) 행 삭제하기

행을 삭제하기 위해서는`drop`메소드를 이용할 수 있습니다.

```python
# a 지우기
df = df.drop(index='a',axis=0)
df
```

## 3. 조건에 따른 Indexing하기 

---

+ `Dataframe`은 numpy array와 마찬가지로, 조건에 따라 indexing을 할 수 있습니다.



### (1) 비교 연산자

비교 연산의 결과는 위와 같은 True/False 값을 가진 Series을 반환합니다.

```python
df.국어 > 80
```

```python
# 결과를 뒤집고 싶으면, ~ 연산자를 이용
~(df.국어 > 80)
```

```python
# 둘다 만족
(df.국어 > 80) & (df.수학 > 80)
```

```python
# 둘 중 하나라도 만족
(df.국어 > 80) | (df.수학 > 80)
```

```python
# 이외에도 값의 포함 유무를 알 수 있는 isin도 있습니다.
df.이름.isin(["호동","윤호"])
```

### (2) Boolean Indexing 적용하기

비교 연산의 결과에서 **참(True)**인 행만을 가져옵니다.

```python
# 국어 성적이 80이 넘는 사람의 것만 가져옴
df.loc[df.국어 > 80]
```

```python
# 국어 성적이 80이 넘는사람들의 이름만 가져오기
df.loc[(df.국어 > 80),"이름"]
```
