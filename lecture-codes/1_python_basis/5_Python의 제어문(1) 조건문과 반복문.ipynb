# \[ 1. 조건문(if) \]
<hr><hr>
> *조건문의 기본 구조는 아래와 같습니다.*


```python
if 조건문1:
    수행할 문장1
    수행할 문장2
    ...
elif 조건문2:
    수행할 문자a
    수행할 문자b
    ...
else:
    수행할 문장A
    수행할 문장B
    ...
```

> + 조건문 1이 만족하면 `수행할 문장1`과 `수행할 문장2`가 수행<br>
> + 조건문 2이 만족하면 `수행할 문장a`과 `수행할 문장b`가 수행<br>
> + 두 조건문 모두 만족하지 않으면 `수행할 문장A`과 `수행할 문장B`가 수행<br>
이 됩니다.

> *조건문에서는 **if만 필수**이고 elif와 else는 없어도 됩니다.*

## 1. 블록문(Block Statement)이란?
<hr>
+ 블록문이란 연속된 코드의 묶음으로, 코드는 다수의 블록문으로 구성되어 있고 블록문 안에 다수의 블록문이 중첩되어 있습니다. 
+ 파이썬에서는 들여쓰기(indent)로 블록을 구분합니다.
+ 블록이 없다면 코드의 연속성이 어떤지 확인할 방법이 없습니다.

**들여쓰기의 차이로 어떤 식으로 동작이 차이나는지 확인해봅시다.**



┌─<i> ⌨️ 예제 코드 </i>────────────────────────┐ <br>
````python
condition_statement = ?? # <- 여기에 True와 False를 각각 넣어봅시다.

if condition_statement:
    print("조건문이 참입니다.")
print("조건문에 해당하는 수행문은 여기까지 입니다.")
````
└────────────────────────────────┘


### (1) Condition_statement가 "참" 일 때

`````python
condition_statement = True

if condition_statement is True:
    print("조건문이 참입니다.")
    
    print("조건문에 해당하는 수행문은 여기까지 입니다.")
`````

`````python
condition_statement = True

if condition_statement:
    print("조건문이 참입니다.")

print("조건문에 해당하는 수행문은 여기까지 입니다.")
`````

> 참일 때는 첫번째 Cell의 결과와 두 번째 Cell의 결과가 동일합니다.

### (2) Condition statement가 "거짓" 일 때

`````python
condition_statement = False

if condition_statement is True:
    print("조건문이 참입니다.")
    
    print("조건문에 해당하는 수행문은 여기까지 입니다.")
`````

`````python
condition_statement = False

if condition_statement:
    print("조건문이 참입니다.")

print("조건문에 해당하는 수행문은 여기까지 입니다.")
`````

###  

> `print("조건문에 해당하는 수행문은 여기까지 입니다.")` 구문은 조건문이 참일 때만 출력되길 원하였지만, <br>
> 아래에서는 indent가 되지 않았기 때문에 조건문이 거짓이었음에도 불구하고 출력되게 되었습니다.<br>
> Python의 indent에 따라 어디까지 연속된 수행문인지를 판단할 수 있습니다.

## 2. 조건문에서의 조건문( Condition Statement )
<hr>
+ 파이썬 조건문 if에서는 단순히 `True`, `False`만이 아니라, 다양한 값들을 `True`와 `False`와 같이 이용할 수 있습니다.


### (1) Boolean에서의 참 거짓

가장 기본적인 조건문입니다.

`````python
if True:
    print("참입니다.")
`````

`````python
if False: # 실행되지 않음
    print("거짓입니다.")
`````

`````python
# 주로 비교 연산을 통해 할 때, Boolean으로 동작합니다.

x = 5
y = 2
if x > y: # 참값
    print("x가 y보다 큽니다.")
else:
    print("x가 y보다 작습니다.")
`````

### (2) 숫자에서의 참거짓

숫자는 0을 제외하고는 모두 참으로 인정됩니다.

`````python
if 1:
    print("1은 통과합니다.")
    
if 0:
    # 거짓으로 아래는 출력하지 못합니다.
    print("0은 통과하지 못합니다.")

if -3:
    print("-3은 통과합니다.")
`````

### (3) String에서의 참 거짓
빈 문자열을 제외하고 모두 참으로 인정됩니다.

`````python
if "hello~":
    print("참입니다!")
`````

`````python
if "":
    print("거짓입니다")
`````

위의 식은 사실 아래와 같습니다. <br>
파이썬은 보다 가독성이 좋은 코드를 추구하므로 위와 같이 서술합니다.

`````python
string = ""

if len(string) != 0:
    print("거짓입니다.")
`````

### (4) 리스트에서의 참 거짓

빈 리스트을 제외하고 모두 참으로 인정됩니다.

`````python
lst = [1,2,3]

if lst:
    print("참입니다!")
`````

`````python
lst = []

if lst:
    print("거짓입니다!")
`````

string과 마찬가지로 사실 이 코드는 아래와 같습니다.

`````python
lst = []

if len(lst) != 0:
    print("거짓입니다!")
`````

### 예시 1.) 돈으로 물건 구매하기

┌─<i> ✍︎ Example </i>───────────────────────────────────┐ <br>

가진 돈에 따라 편의점에서 어떤 물건을 구매할 수 있는지를 알려주는 코드를 만들어봅시다. <br>

└──────────────────────────────────────────┘

`````python
money = 2000
if money >= 3800:
    print("콜라를 구매할 수 있습니다.")
elif money >=1500:
    print("과자를 구매할 수 있습니다.")
elif money >= 600:
    print("계란을 구매할 수 있습니다.")
`````

> 만족하는 조건문은 `money >= 1500`와 `money >= 600` 모두이지만<br>
money >=1500이 먼저 등장했기 때문에, "`과자를 구매할 수 있습니다.`"가 출력되었습니다.

`````python
money = 2000
if money >= 3800:
    print("콜라를 구매할 수 있습니다.")
if money >=1500:
    print("과자를 구매할 수 있습니다.")
if money >= 600:
    print("계란을 구매할 수 있습니다.")
`````

> elif로 조건문이 연결되어 있는 경우에는 "조건문에 만족한 경우가 있는 경우"와 "해당 경우"만 수행하고 **이후 조건문은 수행하지 않습니다.** <br>
> 이후 조건문도 모두 동등하게 수행하길 원한다면 if로 코드를 짜야 합니다.

```python

```

> *while문은 조건문이 참인 동안에 while 문 아래에 속하는 문장들이 반복해서 수행됩니다.*

## 1. 반복문 살펴보기
<hr>

### 예시 3.) 10부터 1까지 출력하는 카운터

┌─<i> ✍︎ Example  </i>───────────────────────────────────┐ <br>

10부터 1까지 차례대로 하나씩 줄어들면서 그 값을 출력하는 카운터 프로그램을 만들어 봅시다. <br> 

└──────────────────────────────────────────┘




`````python
counter = 10
while counter > 0:
    print(counter)
    counter = counter - 1
`````

### (1) 반복문에서의 조건문

if 조건문과 같이 꼭 True/False값이 아니어도 됩니다.

`````python
a = [1,2,3,4,5]

while a:
    print(a.pop())
`````

`````python
a = 5

while a:
    print(a)
    a = a- 1
`````

### (2) 무한루프

계속 참인 경우에는, 무한 루프라는 현상이 발생하여 파이썬 프로그램이 해당 연산만 계속해서 끝나지 않게 됩니다.<br>
이럴 경우, 강제로 해당 Cell을 정지시켜주어야 합니다.

`````python
while True:
    # 조건문이 계속 참이므로, 무한 루프 상태에 빠지게 됨
    print("hello!!!")
`````

### (3) 반복문에서의 `break`와 `continue`

위와 같은 상태에서 반복문에서 빠져나가기 위해서는, `break` 구문을 이용하면 됩니다.

`````python
count = 5

while True:
    # 조건문이 계속 참이므로, 무한 루프 상태에 빠지게 됨
    print("hello!!!")
    count = count - 1 # count가 하나씩 차감됨
    if count == 0:
        # count가 0이 되는 순간 break로 반복문 탈출
        break
`````

* `continue`는 이후 실행할 구문을 건너뛰게 해줍니다.

`````python
count = 5

while count:
    count = count - 1 # count가 하나씩 차감됨
    
    if count % 2 == 0:
        # count가 홀수 일때는 출력 생략
        continue
    print("counter : ", count)
`````

### 예시 4.) 음료수 행사 기계

┌─<i> ✍︎ Example  </i>───────────────────────────────────┐ <br>

100개만큼 재고가 있는 음료수 기계가 있습니다. <br>
이 기계는 사람들이 원하는 만큼 음료수를 줍니다.<br>
그 기계에 들어갈 프로그램을 한번 짜봅시다. <br>

└──────────────────────────────────────────┘


`````python
s = input("입력한 숫자만큼 음료수를 드립니다.")
`````

`````python
type(s)
`````

`````python
type(int(s))
`````

`````python
num_drink = 100

while True:
    # 사용자로부터 입력을 받는 부분
    purchase = int(input("입력한 숫자만큼 음료수를 드립니다."))
    
    if num_drink - purchase < 0:
        # 재고보다 요청한 양이 많으면 거래 종료
        print("재고가 바닥났습니다. 거래를 종료합니다.")
        break
    elif num_drink - purchase == 0:
        # 남은 재고만큼 요청하였을 때에는 마지막 출력 생략
        print("{} 만큼 음료수를 드립니다.".format(num_drink))
        num_drink = num_drink - purchase
        continue
    else:
        # 재고보다 요청한 양이 적으면 제공
        print("{} 만큼 음료수를 드립니다.".format(num_drink))
        num_drink = num_drink - purchase
    
    print("음료수가 {}개 남았습니다.".format(num_drink))
`````
