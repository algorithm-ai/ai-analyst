조건 :

- +, - , *, / 은 모두 magic method 로 만드세요.

```python
# code here

class Calculator(object) :
  def __init__(self,x):
    self.x = x
  def __add__(self, other):
    add = self.x + other.x
    return add
  def __sub__(self, other):
    sub = self.x - other.x
    return sub
  def __mul__(self, other):
    mul = self.x * other.x
    return mul
  def __truediv__(self, other):
    div = self.x / other.x
    return div
```

# Q5

"OOXXOXXOOO"와 같은 OX퀴즈의 결과가 있다.<br>
 O는 문제를 맞은 것이고, X는 문제를 틀린 것이다. <br>
 문제를 맞은 경우 그 문제의 점수는 그 문제까지 연속된 O의 개수가 된다. <br><br>

 "OOXXOXXOOO"의 점수는 1+2+0+0+1+0+0+1+2+3 = 10점이다.<br>
OX퀴즈의 결과가 주어졌을 때, 점수를 구하는 함수(calc_score)를 작성하시오.<br><br>

입력

OOXXOXXOOO

출력 

10

입력

OOXXOOXXOO

출력 

9

입력

OXOXOXOXOXOXOX

출력 

7

입력

OOOOOOOOOO

출력 

55

입력

OOOOXOOOOXOOOOX

출력 

30

```python
# code here
def calc_score(str):
  cnt = 0 # 갯수
  sum = 0 #합
  for s in str:
    if s == "O": 
      cnt +=1 # 0이면 하나씩 늘리면서 더함
    elif s == "X":
      cnt = 0 # 초기화
    sum += cnt
  return sum
```

```python
def calc_score(s):
  def sigma(n):  #함수내에서만 사용가능
    return int((n*(n+1))/2)
    
  l = s.split("X")
  print(l)
  
  lengths = list(map(len,l))
  print(lengths)
  
  scores = list(map(sigma, lengths))
  print(scores)
  
  return sum(scores)
```

```python
inputs = ['OOXXOXXOOO', 'OOXXOOXXOO', 'OXOXOXOXOXOXOX', 'OOOOOOOOOO', 'OOOOXOOOOXOOOOX']
outputs = [10, 9, 7, 55, 30]

if list(map(calc_score, inputs)) == outputs:
    print("정답입니다!!!!")
else:
    raise ValueError("다시 확인해주세요!")
```

문제<br>
리스트 l과 정수 x가 주어진다. 이때, l에서 x보다 작은 수를 모두 출력하는 함수(print_less)를 작성하시오.<br>

입력<br>

l : [ 1 10 4 9 2 3 8 5 7 6 ]

x : 5

출력<br>

[1 4 2 3 ]

```python
# code here
def print_less(lst,x):
  ans = [] # 정답이 리스트형이라 리스트 지정
  for s in lst:
    if s < x:
      #print(s)
      ans.append(s)
  return ans


l = [1,10,4,9,2,3,8,5,7,6]
x = 5

print_less(l,x)
```

문제

어떤 양의 정수 X의 자리수가 등차수열을 이룬다면, 그 수를 한수라고 한다. 등차수열은 연속된 두 개의 수의 차이가 일정한 수열을 말한다. N이 주어졌을 때, 1보다 크거나 같고, N보다 작거나 같은 한수의 개수를 출력하는 프로그램을 작성하시오.

입력

첫째 줄에 1,000보다 작거나 같은 자연수 N이 주어진다.

출력
첫째 줄에 1보다 크거나 같고, N보다 작거나 같은 한수의 개수를 출력한다.

입력

110

출력

99


```python
# code here
def hansu(num):
  ans_8 = 0 #초기화
  if num > 1000:
    print("1000보다 큰수는 입력할 수 없습니다.")
    return
  
  list_cnt = []
  for numb in str(num):
    list_cnt.append(int(numb))
    print(len(list_cnt))
  
  if len(list_cnt) <= 2: #두자리 미만수는 어차피 다 포함됨
    ans_8 = num
  
  if len(list_cnt) >= 3:
    ans_8 = 99 # 두자리 수까지의 답 99개
    for num_tmp in range(100,num+1): #반복횟수 100부터 num까지
      list_tmp = []
      for k in str(num_tmp):
        list_tmp.append(int(k)) #하나씩 int형으로 대입
      a = (list_tmp[1] - list_tmp[0])
      b = (list_tmp[2] - list_tmp[1])
      if a == b :
        ans_8 +=1
      
    
  return ans_8
      
      
```
