# Q4

영어 대소문자와 띄어쓰기만으로 이루어진 문자열이 주어진다. 이 문자열에는 몇 개의 단어가 있을까? 이를 구하는 함수(count_words)를 작성하시오. 

입력

The Curious Case of Benjamin Button 

출력

6

입력

Mazatneunde Wae Teullyeoyo

출력

3

입력

Teullinika Teullyeotzi

출력

2

```
# code here
l = ['The Curious Case of Benjamin Button', 'Mazatneunde Wae Teullyeoyo', 'Teullinika Teullyeotzi']

def count_words(l) :
    a = dict()
    b = list()
    for idx, name in enumerate(l,1):
        a[idx] = name
        aa = str(a[idx])
        aa = aa.split(" ")
        aa = len(aa)
        b.append(aa)
    return b
count_words(l)
```

```
l = ['The Curious Case of Benjamin Button', 'Mazatneunde Wae Teullyeoyo', 'Teullinika Teullyeotzi']
if list(map(count_words, l)) == [6, 3, 2]:
    print("정답입니다!!!!")
else:
    raise ValueError("다시 확인해주세요!")
```

# Q5

"OOXXOXXOOO"와 같은 OX퀴즈의 결과가 있다.<br>
 O는 문제를 맞은 것이고, X는 문제를 틀린 것이다. <br>
 문제를 맞은 경우 그 문제의 점수는 그 문제까지 연속된 O의 개수가 된다. <br><br>

 "OOXXOXXOOO"의 점수는 1+2+0+0+1+0+0+1+2+3 = 10점이다.<br>
OX퀴즈의 결과가 주어졌을 때, 점수를 구하는 함수(calc_score)를 작성하시오.<br><br>

입력

OOXXOXXOOO

출력 

10

입력

OOXXOOXXOO

출력 

9

입력

OXOXOXOXOXOXOX

출력 

7

입력

OOOOOOOOOO

출력 

55

입력

OOOOXOOOOXOOOOX

출력 

30

```
# code here

def sigma(n) : #시그마 함수 정의    
    return n*(n+1)/2

a = 'OOXXOXXOOO'

def calc_score(a) :
    a = a.split('X')
    a =list(map(len,a))
    c = list(map(sigma,a))
      
    return sum(c)

calc_score(a)

```

```
inputs = ['OOXXOXXOOO', 'OOXXOOXXOO', 'OXOXOXOXOXOXOX', 'OOOOOOOOOO', 'OOOOXOOOOXOOOOX']
outputs = [10, 9, 7, 55, 30]

if list(map(calc_score, inputs)) == outputs:
    print("정답입니다!!!!")
else:
    raise ValueError("다시 확인해주세요!")
```

입력<br>

l : [ 1 10 4 9 2 3 8 5 7 6 ]

x : 5

출력<br>

[1 4 2 3 ]

```
# code here
l = [ 1, 10, 4, 9, 2, 3, 8, 5, 7, 6 ]
x = 5

def print_less(a) :
    for i in range(0,len(l)) :
        if a[i] < x :
            print(a[i])
print_less(l)            
```

문제

배열을 정렬하는 것은 쉽다. 수가 주어지면, 그 수의 각 자리수를 내림차순으로 정렬해보자.

입력

첫째 줄에 정렬하고자하는 수 N이 주어진다. N은 1,000,000,000보다 작거나 같은 자연수이다.

출력

첫째 줄에 자리수를 내림차순으로 정렬한 수를 출력한다.

입력

'2143'

출력 <br>

'4321' <br>

```
# code here
def array_num(a) : 
    a = list(str(a))
    a.sort(reverse=True)
    a = int("".join(a))
    return a
array_num(2143)
```
