# Q3

문제 두 자연수 A와 B가 주어진다. 이때, A+B, A-B, A*B, A/B(몫), A%B(나머지)를 계산하는 Class를 작성하시오.


조건 : 
```python3
class Calculator:
        def __init__(self, number):
                self.number = number
```
를 선언 해주세요.

- +, - , *, / 은 모두 magic method 로 만드세요.

````python
# code here
class Calculator(object):
    def __add__(a,b):
        print('A+B = {}'.format(a+b))
    def __sub__(a,b):
        print('A-B = {}'.format(a-b))
    def __mul__(a,b):
        print('A*B = {}'.format(a*b))
    def __div__(a,b):
        print('A/B = {}'.format(a/b))
    def __mod__(a,b):
        print('A%B = {}'.format(a%b))   
      
````

````python
assert 'Calculator' in globals(), "클래스 명을 Calculator로 해주세요"

a = 5
b = 2

x = Calculator(5)
y = Calculator(2)
_add = x + y
_sub = x - y
_mul = x * y
_truediv = x / y
_mod = x % y

assert _add == a+b,(
    f"{a+b}이 되어야 합니다.")
assert _sub == a-b,(
    f"{a-b}이 되어야 합니다.")
assert _mul == a*b,(
    f"{a*b}이 되어야 합니다.")
assert _truediv == a/b,(
    f"{a/b}이 되어야 합니다.")
assert _mod == a%b,(
    f"{a%b}이 되어야 합니다.")


print("정답입니다!!")
````

# Q5

"OOXXOXXOOO"와 같은 OX퀴즈의 결과가 있다.<br>
 O는 문제를 맞은 것이고, X는 문제를 틀린 것이다. <br>
 문제를 맞은 경우 그 문제의 점수는 그 문제까지 연속된 O의 개수가 된다. <br><br>

 "OOXXOXXOOO"의 점수는 1+2+0+0+1+0+0+1+2+3 = 10점이다.<br>
OX퀴즈의 결과가 주어졌을 때, 점수를 구하는 함수(calc_score)를 작성하시오.<br><br>

입력

OOXXOXXOOO

출력 

10

입력

OOXXOOXXOO

출력 

9

입력

OXOXOXOXOXOXOX

출력 

7

입력

OOOOOOOOOO

출력 

55

입력

OOOOXOOOOXOOOOX

출력 

30

````python
# code here
def calc_score(result):
  score = 0
  final_score = 0
  for i in range(0,len(result)):
    if result[i]== "O":
      score = score + 1
      final_score = final_score + score 
    else :
      score = 0
    i = i+1  
  return final_score 
      
  
````

````python
inputs = ['OOXXOXXOOO', 'OOXXOOXXOO', 'OXOXOXOXOXOXOX', 'OOOOOOOOOO', 'OOOOXOOOOXOOOOX']
outputs = [10, 9, 7, 55, 30]

if list(map(calc_score, inputs)) == outputs:
    print("정답입니다!!!!")
else:
    raise ValueError("다시 확인해주세요!")
````

문제<br>
리스트 l과 정수 x가 주어진다. 이때, l에서 x보다 작은 수를 모두 출력하는 함수(print_less)를 작성하시오.<br>

입력<br>

l : [ 1 10 4 9 2 3 8 5 7 6 ]

x : 5

출력<br>

[1 4 2 3 ]

````python
# code here
def print_less(I, x):
  list = []
  for i in range(0, len(I)):
    if I[i] < x :
      list.append(I[i])
      i = i+1 
  return list
````

````python
I = list(map(int, input().split()))
x  = int(input())

print_less(I, x)
````
