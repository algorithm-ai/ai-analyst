## 문제 3. 변수 카운터 만들기

+ 초기값은 0이고, `update_op`를 실행시킬 때마다 1씩 올라가는 변수를 만들어 주세요.

`````python
# 우선 실행해 주세요
tf.reset_default_graph()
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* 변수명은 `answer3`로, 값을 1씩 올려주는 Operation의 이름은 `update_op`로 해주세요

`````python
# 정답을 입력해 주세요
answer3 = tf.Variable(0)
update_op = tf.assign_add(answer3, 1, name='update_op')

`````

#### 정답 확인
* 아래와 같이 출력이 나오도록 해주세요

![Imgur](https://i.imgur.com/1UBgG2w.png)

`````python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for i in range(10):
    sess.run(update_op)
    print("{}번째 실행 : {}".format(i,sess.run(answer3)))
`````
