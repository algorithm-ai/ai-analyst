## 문제 1. odds ratio으로 바꾸기 

텐서플로우를 이용해, 확률(p)을 odds ratio의 값으로 바꾸어주세요.

````
p = 0.01
````

`````python
# 우선 실행해주세요
p = 0.01
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer1`에 답을 입력해주세요 


`````python
# 정답을 입력해주세요
tf.reset_default_graph()
graph = tf.get_default_graph()

p = tf.constant(p)
odds = tf.divide(p, (1-p))

with tf.Session() as sess:
    # 변수 초기화하기
    sess.run(tf.global_variables_initializer())
    print(sess.run(odds))
`````

#### 정답 확인
정답 : 0.01010101 

## 문제 2 .logits 계산하기

텐서플로우를 이용해, 확률(p)을 logits으로 바꾸어주세요.

````
p = 1/3
````

`````python
# 우선 실행해 주세요.
p = 1/3
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer2`에 답을 입력해 주세요


`````python
tf.reset_default_graph()
graph = tf.get_default_graph()

p = tf.constant(p)
odds = tf.divide(p, (1-p))
logits = tf.log(odds)

sess = tf.Session(graph=graph)
# 변수 초기화하기
sess.run(tf.global_variables_initializer())
sess.run(logits)
`````

## 문제 3. logitstic 계산하기

텐서플로우를 이용해 확률(p)을 logitstic으로 바꾸어주세요.

`````python
# 우선 실행해 주세요.
p =  0.7
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer3`에 답을 입력해 주세요

`````python
tf.reset_default_graph()
graph = tf.get_default_graph()

p = tf.constant(p)
# odds = tf.divide(p, (1-p))
# logits = tf.log(odds)

# logistic = 1/(1+tf.exp(-logits))
logistic = 1/(1+tf.exp(-p))
# logistic = tf.divide(1, 1 + tf.exp(-logits))

sess = tf.Session(graph=graph)
# 변수 초기화하기
sess.run(tf.global_variables_initializer())
sess.run(logistic)
`````

## 문제 4. Logistic gradient 계산하기 

텐서플로우를 이용해 -10부터 10까지  0.1 단위로 logitstic 
함수를 미분해 시각화 해보세요



`````python
# 우선 실행해 주세요.
ps = np.arange(-10, 10, 0.1)
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요


`````python
tf.reset_default_graph()
graph = tf.get_default_graph()

ps = tf.constant(ps)
# odds = tf.divide(p, (1-p))
# logits = tf.log(odds)

# logistic = 1/(1+tf.exp(-logits))
logistic = 1/(1+tf.exp(-ps))
dsig = logistic * (1-logistic )
# logistic = tf.divide(1, 1 + tf.exp(-logits))

sess = tf.Session(graph=graph)
# 변수 초기화하기
sess.run(tf.global_variables_initializer())
dsigs = sess.run(dsig)

# 그리기
plt.plot(dsigs)
plt.show()
`````

## 문제 5. 엔트로피 계산하기

오늘 일어날 사건은 A, B, C, D 4가지 입니다. 

그리고 각 일어날 확률은 아래와 같습니다. 


A 사건은 일어날 확률이 0.7 입니다. <br>
B 사건은 일어날 확률이 0.26 입니다. <br>
C 사건은 일어날 확률이 0.02 입니다. <br>
D 사건은 일어날 확률이 0.02 입니다. <br>

오늘 일어날 사건의 엔트로피를 텐서플로우를 이용해 구해주세요.

`````python
# 우선 실행해 주세요.
ps = [0.7, 0.26, 0.02, 0.02]
`````

#### 답안지 입력란
* 아래에 답을 서술하여 주세요.
* `answer5`에 답을 입력해 주세요.


`````python
tf.reset_default_graph()
graph = tf.get_default_graph()

ps = tf.constant(ps)

entropy = tf.reduce_sum(-ps * tf.log(ps))

sess = tf.Session(graph=graph)
# 변수 초기화하기
sess.run(tf.global_variables_initializer())
sess.run(entropy)
`````

#### 정답 확인 

정답 :  0.7563925


## 문제 6. Cross - Entropy 계산하기

아래 경우 Cross-entropy 을 구해주세요.

|정답  | 0   | 1   |
|-----|-----|-----|
|예측값| 0.3 | 0.7 |

#### 답안지 입력란
* 아래에 답을 서술하여 주세요
* `answer6`에 답을 입력해 주세요.


`````python
tf.reset_default_graph()
graph = tf.get_default_graph()

cross_entropy = -(1*tf.log(0.7))

sess = tf.Session(graph=graph)
# 변수 초기화하기
sess.run(tf.global_variables_initializer())
sess.run(cross_entropy)
`````

#### 정답 확인 

정답 :  0.35667497
